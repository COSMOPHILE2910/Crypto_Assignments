```python
import hashlib
import os

def get_truncated_hash(message_bytes):
    """Generates SHA-256 and truncates it to the first 5 bytes (40 bits)."""
    full_hash = hashlib.sha256(message_bytes).digest()
    return full_hash[:5]  # 5 bytes = 40 bits

def find_collision():
    seen_hashes = {}
    attempts = 0

    print("Searching for a 40-bit collision... (This usually takes a few seconds)")

    while True:
        attempts += 1
        
        # 1. Generate a random 16-byte message
        random_message = os.urandom(16)
        
        # 2. Hash and truncate to 40-bit
        truncated_hash = get_truncated_hash(random_message)

        # 3. Check if we've seen this 40-bit hash before
        if truncated_hash in seen_hashes:
            match_message = seen_hashes[truncated_hash]
            
            # Make sure it's not the exact same message by pure chance
            if match_message != random_message:
                print(f"\n✅ Collision found after {attempts:,} attempts!")
                return random_message, match_message, truncated_hash

        # 4. If not seen, store it and keep going
        seen_hashes[truncated_hash] = random_message

# Run the attack
msg1, msg2, col_hash = find_collision()

print("\n--- RESULTS ---")
print(f"Message 1 (Hex): {msg1.hex()}")
print(f"Message 2 (Hex): {msg2.hex()}")
print(f"40-bit Hash (Hex): {col_hash.hex()}")
```
