```python
from Crypto.Cipher import AES
from Crypto.Util.strxor import strxor 

def verify_diffusion():
    # 1. Setup - 128-bit key and two inputs with 1-byte difference
    key = b'\x00' * 16
    p1 = b'\x00' * 16
    p2 = b'\x01' + b'\x00' * 15 # Change only the first byte

    # 2. Encryption
    cipher = AES.new(key, AES.MODE_ECB)
    c1 = cipher.encrypt(p1)
    c2 = cipher.encrypt(p2)

    # 3. Difference Analysis (XOR)
    diff = strxor(c1, c2)
    
    print("--- AES Diffusion Result ---")
    print(f"XOR Difference (Hex): {diff.hex()}")
    
    # Count how many bytes were affected (should be 16)
    changed = sum(1 for b in diff if b != 0)
    print(f"Bytes affected: {changed}/16")

if __name__ == "__main__":
    verify_diffusion()
```
